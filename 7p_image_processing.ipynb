```
# There are 2 ways to manage image sizes 
# 1 Either resize from source itself
# 2 create images of same size from source itself

import cv2
import numpy as np
from google.colab.patches import cv2_imshow # Only on Google Colab

i1 = cv2.imread('ownimage1.png')
i2 = cv2.imread('ownimage2.png')

i1_resized = cv2.resize(i1, (100,100))
i2_resized = cv2.resize(i2, (100,100))


newimage = cv2.subtract(i2_resized,i1_resized) # i1-i2 -> ORDER MATTERS 
cv2_imshow(i1_resized)
cv2_imshow(i2_resized)
cv2_imshow(newimage)
```
