```
# How to create dependency trees ?
import spacy
nlp = spacy.load('en_core_web_sm')
# text="Hello world, where are you?"
text="Bhaskar is eating a sandwich"
tokens=nlp(text)

from spacy import displacy # dependency trees
displacy.render(tokens,style='dep',jupyter=True,options={'distance':100})
```

```
# NOUN Phrases are also visible...
for entity in tokens.ents: 
  if(entity.label_=='NORP' or entity.label_=='DATE'):
    print(entity.text+":" +entity.label_)
```

```
# Censoring Contenct through Filters
import spacy
nlp = spacy.load('en_core_web_sm')

# create 2 Functions 
# 1) to clean individual token (lambda) 2) two swipe through (map)

# 1. lambda val: '<CENSORED>' if val.ent_type=='PERSON' else val.string
def censor(token):
  #print(token.ent_type)
  if token.ent_type_ == "PERSON":
    return '<CENSORED>'
  else:
    return token.string

# 2.
def scrub(text):
  doc = nlp(text)
  for ent in doc.ents:
    ent.merge() # this will combine all entities into a single DataStructure that we will map to 
  tokens = map(censor, doc)
  return "".join(tokens)


# sample input

s = """In 1940-50 Alan Turing and John Nash evolved the math and game theory"""
print(scrub(s))
```
