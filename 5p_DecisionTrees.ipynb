```
import numpy as np
import pandas as pd
data =pd.read_csv('data.csv')
data.head()
```

```
df=data.iloc[:,[1,2,5,6,9]] #dynamically taken the columns
df.head()
```

```
y=df.pop('diagnosis') # Remove the feature from df and put into y
```

```
x=df # all my input is x and all my output is y.
```

```
# y = f(x) -> equation for DecisionTree 
from sklearn.model_selection import train_test_split
xtrain,xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=42)
```

```
# Decision Tree, Random Forest -> work on relative comparison
# Normalizing and Not Normalizing data may not cause any difference at all 
```

```
# Below Depth was 4
```

image.png

```
# Weights and Biases, y = Weights * inputs + bias -> is called Machine Learning
```

```
# Decision Tree doesn't help with calculating the above but it helps finding thresholds within which features take decisions 
```

```
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier(max_depth=5)
dtree_light =  DecisionTreeClassifier(max_depth=3)
dtree_deeper =  DecisionTreeClassifier(max_depth=10)
dtree.fit(xtrain, ytrain)
dtree_light.fit(xtrain, ytrain)
dtree_deeper.fit(xtrain, ytrain)
```

```
pred1 = dtree.predict(xtest)
pred2 = dtree_light.predict(xtest)
pred3 = dtree_deeper.predict(xtest)
```

```
# accuracy_score(a, b) = (len(any) - mismatches) / len(any)
from sklearn.metrics import accuracy_score
# Comparing the Trees to each other, not to the actual answer (ytest)
score = accuracy_score(pred1, pred2)
print(score)
score = accuracy_score(pred1, pred3)
print(score)
score = accuracy_score(pred3, pred2)
print(score)
```
